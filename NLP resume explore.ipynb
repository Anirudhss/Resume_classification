```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS as stopword
from string import punctuation
print(stopword)
punctuation=punctuation + '\n' + '—'
print(punctuation)
maintext =''.join(c if c not in map(str,punctuation) else "" for c in resmain)
maintext1 =''.join(c if c not in map(str,range(0,10)) else "" for c in maintext)
maintext1=maintext1.lower()
maintext2 = " ".join(maintext1.split())
print(maintext2)
```

```python
nlp = spacy.load('en_core_web_sm')
maindoc = nlp(maintext2)
tokens = [token.text for token in maindoc]
print(tokens)
```

```python
word_frequencies = {}
for word in maindoc:
    if word.text not in stopword:
        if word.text not in stopset:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
print(word_frequencies)
```

```python
print(len(tokens))
words23=list(word_frequencies.keys())
print(words23)
```

```python
from nltk.stem import wordnet 
from nltk.stem import WordNetLemmatizer
word_lem=WordNetLemmatizer()
mainword=[]
for wordss in words23:
  mainword.append(word_lem.lemmatize(wordss))
print(mainword)  
```
